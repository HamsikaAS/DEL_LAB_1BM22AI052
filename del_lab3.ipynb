```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("/content/heart.csv")
df.head()
```

```python
df1 = df.sample(n = 20 , replace = False)
df1.head()
```

```python
def sigmoid(x):
  return (1/(1+np.exp(-x)))

def tanh(x):
  return np.tanh(x)
```

```python
x = np.array(df1.iloc[:,:-1])
w = np.array([0.05]*len(x[0])).T
h = []
bias = -30
for i in range (len(x)):
  h.append(np.sum(x[i]*w)+ bias)

yp = []
for i in h:
  yp.append(sigmoid(i))

y_pred = []
for i in yp:
  if i>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

y_true = list(df1.iloc[:,-1])
confusion_matrix(y_true, y_pred)
```
