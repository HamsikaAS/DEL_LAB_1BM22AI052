```python
x1 = int(input('enter the 1st input'))
x2 = int(input('enter the 2nd input'))
```

```python
def ReLu(x):
  if x>=0:
    return x
  else:
    return 0

def tan_h(x):
  return np.tanh(x)
```

```python
i = np.matrix([x1,x2])
w = np.matrix([0.5,0.5])
h = i*w.T
h = ReLu(h)
o = tan_h(h)
print(bool(o))
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from numpy import tanh

def softmax(r):
    e = np.exp(r)
    return e / np.sum(e)

# Load Iris dataset
iris = load_iris()

data1 = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                     columns=iris['feature_names'] + ['target'])

df = data1.drop('target', axis=1)
weight_input = np.array([[0.5] * 4] * 3)

hidden = np.dot(df, weight_input.T)

hidden_act = tanh(hidden)

weight_output = np.array([[0.5] * 3] * 3)

output = np.dot(hidden_act, weight_output.T)

l = []
for i in range(len(output)):
    l.append(softmax(output[i]))

print(l)
```
